# REDSHIFT DISTRIBUTION OF GALAXIES

This notebook shows different filters for the DES's redshift distribution of galaxies.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.special as sc
from astropy.cosmology import FlatLambdaCDM
```

```python
import sys
```

## Load DES-Y1 quantities

* Metadata

```python
nbin = 4
bin_a, bin_b = np.tril_indices(nbin)
bin_a += 1
bin_b += 1
```

```python
nbinl = 5
binl_a, binl_b = np.tril_indices(nbinl)
binl_a += 1
binl_b += 1
```

* Cosmology

```python
cosmo = {}
with open('../data/des-y1-test/cosmological_parameters/values.txt') as cosmo_values:
    for line in cosmo_values:
        if line:
            key, val = line.partition('=')[::2]
            cosmo[key.strip()] = float(val)
```

```python
cosmo_astropy = FlatLambdaCDM(H0=cosmo['hubble'], Ob0=cosmo['omega_b'], Om0= cosmo['omega_m'], Tcmb0=2.7)
```

* Distance functions

```python
zdM = np.loadtxt('../data/des-y1-test/distances/z.txt')
dM = np.loadtxt('../data/des-y1-test/distances/d_m.txt')
```

* Redshift distributions

```python
zn = np.loadtxt('../data/des-y1-test/nz_source/z.txt')
nz = [np.loadtxt('../data/des-y1-test/nz_source/bin_%d.txt' % i) for i in range(1, nbin+1)]
```

```python
xf = np.interp(zn, zdM, dM)
```

```python
nlz = [np.loadtxt('../data/des-y1-test/nz_lens/bin_%d.txt' % i) for i in range(1, nbinl+1)]
```

```python
fig, (ax0, ax1) = plt.subplots(ncols=2, constrained_layout=True, figsize=(10,4))

ax0.set_title('Source galaxies')
for nn in nz:
    ax0.plot(xf, nn)
    ax0.set_xlabel(r'Comoving distance, $x$ $(Mpc)$')
    ax0.set_ylabel('Redshift distribution of galaxies')

ax1.set_title('Lens galaxies')
for nn in nlz:
    plt.plot(xf, nn)
    ax1.set_xlabel(r'Comoving distance, $x$ $(Mpc)$')
    
plt.savefig('desy1.pdf')

plt.show()
```

### CMB FILTERS

```python
import camb
```

```python
cosmo = FlatLambdaCDM(H0=67.5, Ob0=0.0483, Om0= 0.3174653243323745)
```

```python
sys.path.append("../")
from lens_filter import lensing_efficiency_cmb
```

```python
nz = 100

pars = camb.CAMBparams()
pars.set_cosmology(H0=67.5, ombh2=0.022, omch2=0.122)
pars.InitPower.set_params(ns=0.965)

results = camb.get_background(pars)
xs = results.conformal_time(0) - results.tau_maxvis
```

```python
# filter
xf = np.linspace(0., xs, 400)
zf = results.redshift_at_comoving_radial_distance(xf)
```

```python
qcmb = lensing_efficiency_cmb(xf, xs)
```

```python
plt.plot(xf, qcmb)
plt.xlabel(r'Comoving distance, $x$ $(Mpc)$')
plt.ylabel('Lensing efficiency, $q_{CMB}$')
plt.show()
```

```python
fc = filter_convergence(xf, zf, qcmb, cosmo)
```

```python
plt.plot(xf, fc)
plt.xlabel(r'Comoving distance, $x$ $(Mpc)$')
plt.ylabel('Filter convergence, $f_{\kappa CMB}$')
plt.show()
```
